```python
# ===============================================
# File: Domain_MEA.ipynb
# ===============================================
print("--- Memuat Modul Domain MEA ---")

# ----- Helper Functions (Wajib ada di setiap file domain) -----
def tampilkan_skala():
    """Menampilkan panduan Skala Likert"""
    print("--- Panduan Penilaian (Skala Likert) ---")
    print("1: Sangat Tidak Setuju / Sangat Buruk")
    print("2: Tidak Setuju / Buruk")
    print("3. Netral")
    print("4. Setuju / Baik")
    print("5. Sangat Setuju / Sangat Baik")
    print("------------------------------------------\n")

def dapatkan_skor(pertanyaan):
    """Fungsi helper untuk meminta input skor 1-5 dengan validasi"""
    while True:
        try:
            print(pertanyaan)
            skor = int(input("Masukkan penilaian Anda (1-5): "))
            
            if 1 <= skor <= 5:
                print("\n") # Memberi spasi antar pertanyaan
                return skor # Kembalikan skor jika valid
            else:
                print("Input tidak valid. Harap masukkan angka antara 1 dan 5.")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

def tampilkan_hasil_domain(nama_domain, hasil_skor, rekomendasi):
    """Menampilkan hasil evaluasi untuk satu domain"""
    total_skor = sum(hasil_skor.values())
    jumlah_pertanyaan = len(hasil_skor)
    rata_rata = total_skor / jumlah_pertanyaan
    
    print(f"--- HASIL EVALUASI DOMAIN {nama_domain} ---")
    print("Detail Skor per Proses:")
    for proses, skor in hasil_skor.items():
        print(f"- {proses}: {skor}")
        
    print("\n--- Ringkasan ---")
    print(f"Rata-rata Skor: {rata_rata:.2f} dari 5.00")
    print(f"Rekomendasi: {rekomendasi(rata_rata)}")
    print("---------------------------------\n\n")
# ----- End Helper Functions -----


# ----- Fungsi Evaluasi Utama Domain Ini -----
def evaluasi_mea():
    """Fungsi utama untuk evaluasi domain MEA"""
    print("Memulai Evaluasi Domain: MEA (Monitor, Evaluate and Assess)\n")
    tampilkan_skala()
    hasil_skor = {}

    # --- Pertanyaan-pertanyaan MEA ---
    # (Ini adalah contoh pertanyaan, silakan disesuaikan oleh kelompok)

    hasil_skor["MEA01"] = dapatkan_skor("MEA01: Seberapa baik Tokopedia memantau kinerja platform Marketplace (kecepatan, uptime) dan membandingkannya dengan target?")
    hasil_skor["MEA02"] = dapatkan_skor("MEA02: Bagaimana efektivitas sistem audit internal (kontrol internal) Tokopedia dalam mengevaluasi proses TI?")
    hasil_skor["MEA03"] = dapatkan_skor("MEA03: Sejauh mana Tokopedia memastikan platform Marketplace-nya patuh terhadap regulasi eksternal (misal: UU Perlindungan Data Pribadi)?")
    
    # --- Rekomendasi (Contoh) ---
    def rekomendasi(rata_rata):
        if rata_rata < 3:
            return "Proses pemantauan, evaluasi, dan kepatuhan (compliance) sangat lemah dan berisiko tinggi."
        elif rata_rata < 4:
            return "Proses pemantauan sudah berjalan, namun perlu penguatan pada area audit internal dan kepatuhan regulasi."
        else:
            return "Proses pemantauan, evaluasi, dan kepatuhan sudah sangat matang dan terkelola dengan baik."

    # --- Tampilkan Hasil ---
    tampilkan_hasil_domain("MEA", hasil_skor, rekomendasi)

# ----- JALANKAN FUNGSI -----
# Baris ini akan otomatis berjalan saat dipanggil oleh %run dari MenuUtama.ipynb
evaluasi_mea()
```
