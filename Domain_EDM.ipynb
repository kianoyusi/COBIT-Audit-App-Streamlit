```python
#### ===============================================
# File: Domain_EDM.ipynb
# ===============================================
print("--- Memuat Modul Domain EDM ---")

# ----- Helper Functions (Wajib ada di setiap file domain) -----
def tampilkan_skala():
    print("--- Panduan Penilaian (Skala Likert) ---")
    print("1: Sangat Tidak Setuju / Sangat Buruk")
    print("2: Tidak Setuju / Buruk")
    print("3: Netral")
    print("4: Setuju / Baik")
    print("5: Sangat Setuju / Sangat Baik")
    print("------------------------------------------\n")

def dapatkan_skor(pertanyaan):
    while True:
        try:
            print(pertanyaan)
            skor = int(input("Masukkan penilaian Anda (1-5): "))
            if 1 <= skor <= 5:
                print("\n")
                return skor
            else:
                print("Input tidak valid. Harap masukkan angka antara 1 dan 5.")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

def tampilkan_hasil_domain(nama_domain, hasil_skor, rekomendasi):
    total_skor = sum(hasil_skor.values())
    jumlah_pertanyaan = len(hasil_skor)
    rata_rata = total_skor / jumlah_pertanyaan
    
    print(f"--- HASIL EVALUASI DOMAIN {nama_domain} ---")
    print("Detail Skor per Proses:")
    for proses, skor in hasil_skor.items():
        print(f"- {proses}: {skor}")
    print("\n--- Ringkasan ---")
    print(f"Rata-rata Skor: {rata_rata:.2f} dari 5.00")
    print(f"Rekomendasi: {rekomendasi(rata_rata)}")
    print("---------------------------------\n\n")
# ----- End Helper Functions -----


# ----- Fungsi Evaluasi Utama Domain Ini -----
def evaluasi_edm():
    print("Memulai Evaluasi Domain: EDM (Evaluate, Direct and Monitor)\n")
    tampilkan_skala()
    hasil_skor = {}

    hasil_skor["EDM01"] = dapatkan_skor("EDM01: Sejauh mana dewan direksi Tokopedia memastikan kerangka tata kelola TI (Marketplace) ditetapkan?")
    hasil_skor["EDM02"] = dapatkan_skor("EDM02: Sejauh mana manajemen puncak memastikan bahwa investasi TI (fitur baru Marketplace) memberikan nilai?")
    hasil_skor["EDM03"] = dapatkan_skor("EDM03: Bagaimana penilaian Anda terhadap cara manajemen puncak Tokopedia mengelola risiko TI (keamanan data, downtime)?")
    
    def rekomendasi(rata_rata):
        if rata_rata < 3: return "Tata kelola (governance) TI di level dewan/direksi perlu evaluasi dan perbaikan segera."
        elif rata_rata < 4: return "Tata kelola TI sudah berjalan, namun perlu peningkatan pada pemantauan."
        else: return "Tata kelola TI di level dewan/direksi sudah berjalan sangat baik."

    tampilkan_hasil_domain("EDM", hasil_skor, rekomendasi)

# ----- JALANKAN FUNGSI -----
# Baris ini akan otomatis berjalan saat dipanggil oleh %run
evaluasi_edm()
```
