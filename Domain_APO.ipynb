```python
#### ===============================================
# File: Domain_APO.ipynb
# ===============================================
print("--- Memuat Modul Domain APO ---")

# ----- Helper Functions (Wajib ada di setiap file domain) -----
def tampilkan_skala():
    print("--- Panduan Penilaian (Skala Likert) ---")
    print("1: Sangat Tidak Setuju / Sangat Buruk")
    print("2: Tidak Setuju / Buruk")
    print("3: Netral")
    print("4: Setuju / Baik")
    print("5: Sangat Setuju / Sangat Baik")
    print("------------------------------------------\n")

def dapatkan_skor(pertanyaan):
    while True:
        try:
            print(pertanyaan)
            skor = int(input("Masukkan penilaian Anda (1-5): "))
            if 1 <= skor <= 5:
                print("\n")
                return skor
            else:
                print("Input tidak valid. Harap masukkan angka antara 1 dan 5.")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

def tampilkan_hasil_domain(nama_domain, hasil_skor, rekomendasi):
    total_skor = sum(hasil_skor.values())
    jumlah_pertanyaan = len(hasil_skor)
    rata_rata = total_skor / jumlah_pertanyaan
    
    print(f"--- HASIL EVALUASI DOMAIN {nama_domain} ---")
    print("Detail Skor per Proses:")
    for proses, skor in hasil_skor.items():
        print(f"- {proses}: {skor}")
    print("\n--- Ringkasan ---")
    print(f"Rata-rata Skor: {rata_rata:.2f} dari 5.00")
    print(f"Rekomendasi: {rekomendasi(rata_rata)}")
    print("---------------------------------\n\n")
# ----- End Helper Functions -----


# ----- Fungsi Evaluasi Utama Domain Ini -----
def evaluasi_apo():
    print("Memulai Evaluasi Domain: APO (Align, Plan and Organize)\n")
    tampilkan_skala()
    hasil_skor = {}

    hasil_skor["APO01"] = dapatkan_skor("APO01: Sejauh mana Tokopedia telah menetapkan kerangka kerja tata kelola TI (Marketplace) dengan jelas?")
    hasil_skor["APO02"] = dapatkan_skor("APO02: Apakah strategi TI (Marketplace) Tokopedia sudah selaras dengan strategi bisnis perusahaan?")
    hasil_skor["APO04"] = dapatkan_skor("APO04: Bagaimana efektivitas Tokopedia dalam mengidentifikasi dan menerapkan inovasi (misal: AI, live shopping)?")
    
    def rekomendasi(rata_rata):
        if rata_rata < 3: return "Perlu adanya perbaikan signifikan pada domain APO."
        elif rata_rata < 4: return "Domain APO sudah cukup baik, namun perlu optimasi."
        else: return "Domain APO sudah berjalan dengan sangat baik."

    tampilkan_hasil_domain("APO", hasil_skor, rekomendasi)

# ----- JALANKAN FUNGSI -----
evaluasi_apo()
```
