```python
# ===============================================
# File: Domain_BAI.ipynb
# =Examples/
# ===============================================
print("--- Memuat Modul Domain BAI ---")

# ----- Helper Functions (Wajib ada di setiap file domain) -----
def tampilkan_skala():
    """Menampilkan panduan Skala Likert"""
    print("--- Panduan Penilaian (Skala Likert) ---")
    print("1: Sangat Tidak Setuju / Sangat Buruk")
    print("2: Tidak Setuju / Buruk")
    print("3: Netral")
    print("4: Setuju / Baik")
    print("5: Sangat Setuju / Sangat Baik")
    print("------------------------------------------\n")

def dapatkan_skor(pertanyaan):
    """Fungsi helper untuk meminta input skor 1-5 dengan validasi"""
    while True:
        try:
            print(pertanyaan)
            skor = int(input("Masukkan penilaian Anda (1-5): "))
            
            if 1 <= skor <= 5:
                print("\n") # Memberi spasi antar pertanyaan
                return skor # Kembalikan skor jika valid
            else:
                print("Input tidak valid. Harap masukkan angka antara 1 dan 5.")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

def tampilkan_hasil_domain(nama_domain, hasil_skor, rekomendasi):
    """Menampilkan hasil evaluasi untuk satu domain"""
    total_skor = sum(hasil_skor.values())
    jumlah_pertanyaan = len(hasil_skor)
    rata_rata = total_skor / jumlah_pertanyaan
    
    print(f"--- HASIL EVALUASI DOMAIN {nama_domain} ---")
    print("Detail Skor per Proses:")
    for proses, skor in hasil_skor.items():
        print(f"- {proses}: {skor}")
        
    print("\n--- Ringkasan ---")
    print(f"Rata-rata Skor: {rata_rata:.2f} dari 5.00")
    print(f"Rekomendasi: {rekomendasi(rata_rata)}")
    print("---------------------------------\n\n")
# ----- End Helper Functions -----


# ----- Fungsi Evaluasi Utama Domain Ini -----
def evaluasi_bai():
    """Fungsi utama untuk evaluasi domain BAI"""
    print("Memulai Evaluasi Domain: BAI (Build, Acquire and Implement)\n")
    tampilkan_skala()
    hasil_skor = {}

    # --- Pertanyaan-pertanyaan BAI ---
    # (Ini adalah contoh pertanyaan, silakan disesuaikan oleh kelompok)

    hasil_skor["BAI01"] = dapatkan_skor("BAI01: Bagaimana penilaian Anda terhadap cara Tokopedia mengelola proyek-proyek TI (misal: peluncuran fitur baru Marketplace) agar selesai tepat waktu?")
    hasil_skor["BAI03"] = dapatkan_skor("BAI03: Sejauh mana proses pengembangan (coding) dan desain solusi di Tokopedia sudah terdefinisi dan berkualitas baik?")
    hasil_skor["BAI06"] = dapatkan_skor("BAI06: Bagaimana efektivitas Tokopedia dalam mengelola perubahan (update) pada aplikasi Marketplace agar meminimalkan bug atau gangguan?")
    hasil_skor["BAI07"] = dapatkan_skor("BAI07: Sejauh mana Tokopedia memastikan bahwa fitur/sistem baru telah diuji (testing) secara menyeluruh sebelum diluncurkan?")
    
    # --- Rekomendasi (Contoh) ---
    def rekomendasi(rata_rata):
        if rata_rata < 3:
            return "Proses pengembangan dan implementasi TI (rilis fitur baru) memiliki kelemahan besar dan berisiko tinggi."
        elif rata_rata < 4:
            return "Proses pengembangan TI sudah baik, namun perlu peningkatan pada manajemen perubahan dan testing."
        else:
            return "Proses pengembangan dan implementasi TI sudah matang dan berjalan sangat baik."

    # --- Tampilkan Hasil ---
    tampilkan_hasil_domain("BAI", hasil_skor, rekomendasi)

# ----- JALANKAN FUNGSI -----
# Baris ini akan otomatis berjalan saat dipanggil oleh %run dari MenuUtama.ipynb
evaluasi_bai()
```
