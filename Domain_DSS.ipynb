```python
##### ===============================================
# File: Domain_DSS.ipynb
# ===============================================
print("--- Memuat Modul Domain DSS ---")

# ----- Helper Functions (Wajib ada di setiap file domain) -----
def tampilkan_skala():
    """Menampilkan panduan Skala Likert"""
    print("--- Panduan Penilaian (Skala Likert) ---")
    print("1: Sangat Tidak Setuju / Sangat Buruk")
    print("2: Tidak Setuju / Buruk")
    print("3: Netral")
    print("4: Setuju / Baik")
    print("5: Sangat Setuju / Sangat Baik")
    print("------------------------------------------\n")

def dapatkan_skor(pertanyaan):
    """Fungsi helper untuk meminta input skor 1-5 dengan validasi"""
    while True:
        try:
            print(pertanyaan)
            skor = int(input("Masukkan penilaian Anda (1-5): "))
            
            if 1 <= skor <= 5:
                print("\n") # Memberi spasi antar pertanyaan
                return skor # Kembalikan skor jika valid
            else:
                print("Input tidak valid. Harap masukkan angka antara 1 dan 5.")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")

def tampilkan_hasil_domain(nama_domain, hasil_skor, rekomendasi):
    """Menampilkan hasil evaluasi untuk satu domain"""
    total_skor = sum(hasil_skor.values())
    jumlah_pertanyaan = len(hasil_skor)
    rata_rata = total_skor / jumlah_pertanyaan
    
    print(f"--- HASIL EVALUASI DOMAIN {nama_domain} ---")
    print("Detail Skor per Proses:")
    for proses, skor in hasil_skor.items():
        print(f"- {proses}: {skor}")
        
    print("\n--- Ringkasan ---")
    print(f"Rata-rata Skor: {rata_rata:.2f} dari 5.00")
    print(f"Rekomendasi: {rekomendasi(rata_rata)}")
    print("---------------------------------\n\n")
# ----- End Helper Functions -----


# ----- Fungsi Evaluasi Utama Domain Ini -----
def evaluasi_dss():
    """Fungsi utama untuk evaluasi domain DSS"""
    print("Memulai Evaluasi Domain: DSS (Deliver, Service and Support)\n")
    tampilkan_skala()
    hasil_skor = {}

    # --- Pertanyaan-pertanyaan DSS ---
    # (Ini adalah contoh pertanyaan, silakan disesuaikan oleh kelompok)

    hasil_skor["DSS01"] = dapatkan_skor("DSS01: Bagaimana penilaian Anda terhadap keandalan operasional (ketersediaan server, minim downtime) dari platform Marketplace Tokopedia?")
    hasil_skor["DSS02"] = dapatkan_skor("DSS02: Sejauh mana efektivitas Tokopedia (Customer Service) dalam menanggapi dan menyelesaikan keluhan atau insiden dari pengguna?")
    hasil_skor["DSS03"] = dapatkan_skor("DSS03: Bagaimana efektivitas Tokopedia dalam mengidentifikasi akar masalah (root cause) dari insiden yang berulang (misal: bug aplikasi)?")
    hasil_skor["DSS05"] = dapatkan_skor("DSS05: Bagaimana penilaian Anda terhadap pengelolaan layanan keamanan TI di Tokopedia (perlindungan data pengguna, keamanan transaksi)?")
    
    # --- Rekomendasi (Contoh) ---
    def rekomendasi(rata_rata):
        if rata_rata < 3:
            return "Layanan operasional dan dukungan pelanggan (CS) memiliki masalah serius yang perlu segera ditangani."
        elif rata_rata < 4:
            return "Layanan operasional sudah cukup baik, namun perlu peningkatan pada kecepatan respons insiden."
        else:
            return "Layanan operasional, dukungan pelanggan, dan keamanan sudah berjalan sangat baik dan andal."

    # --- Tampilkan Hasil ---
    tampilkan_hasil_domain("DSS", hasil_skor, rekomendasi)

# ----- JALANKAN FUNGSI -----
# Baris ini akan otomatis berjalan saat dipanggil oleh %run dari MenuUtama.ipynb
evaluasi_dss()
```
